```python
import numpy as np
import pandas as pd
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')
```

## [0] Data Preprocessing


```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/df['population'], 0)
```

```python
df['gdp_per_capita']
```

```python
# Making DataFrame Pipeline interactive
idf = df.interactive()
```

## [0] CO2 emisson over time by continent

```python
# Define Panel widgets
year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2021,step=5, value=1850)
year_slider
```

```python
# Radio button for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['co2','co2_per_capita',],
    button_types = 'success'
)
```

```python
continents = ['World', 'Asia', 'Oceania','Europe','Africa','NA','SA','Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
    
)
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x='year', by='country', y= yaxis_co2, line_width = 2, title = "CO2 Emission By Continent")
co2_plot
```

##  [2]  Table- CO2 Emisson over time by continent

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination = 'remote', page_size = 10, sizing_mode = 'stretch_width')
co2_table
```

## [3] CO2 vs GDP scatterplot

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country','year','gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)

)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita', y='co2',by = 'country' , size =80, kind = 'scatter',
                                                                alpha = 0.7,
                                                                legend = False,
                                                                height = 500 ,
                                                                width = 500)
co2_vs_gdp_scatterplot
```

## [4] Bar chart with CO2 sources by continent

```python
yaxis_co2_sources = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['coal_co2','oil_co2', 'gas_co2'],
    button_types = 'success'
)

contienets_excl_world = ['World', 'Asia', 'Oceania','Europe','Africa','NA','SA','Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(contienets_excl_world))
    ]
    .groupby(['year','country'])[yaxis_co2_sources].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
    
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar',
                                                     x='country',
                                                    y =yaxis_co2_sources,
                                                    title='CO2 source by continent')
co2_source_bar_plot
```

### Creating Dashboard

```python
# Layout using Template
template = pn.template.FastListTemplate(
    title = 'World CO2 emission dashboard',
    sidebar = [pn.pane.Markdown('# CO2 Emisson and Climate Change'),
              pn.pane.Markdown("#### Carbon dioxide emissions from human activities, particularly the burning of fossil fuels, are the primary cause of global climate change and its associated impacts."),
              pn.pane.PNG('climate.png', sizing_mode = 'scale_both'),
              pn.pane.Markdown('## Settings'),
              year_slider],
    main = [pn.Row(pn.Column(yaxis_co2, co2_plot.panel(width = 650 ),margin = (0,25)),
                             co2_table.panel(width=500)),
            pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=500),margin=(0,25)),
                   pn.Column(yaxis_co2_sources, co2_source_bar_plot.panel(width=550)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
```

```python
template.show()
```
